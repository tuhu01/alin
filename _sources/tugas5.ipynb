## Eigen Value dan Eigen Vector

**Definisi**  
Jika A adalah sebuah matriks, maka vektor tidak nol x dinamakan vector eigen dari A jika Ax adalah kelipatan skalar dari x yakni

$$Ax = \lambda x$$  
skalar $\lambda$ dinamakan nilai eigen dari A dan X dikatakan vektor eigen yang bersesuaian dengan $\lambda$

Adapun syarat bahwa vektor eigen adalah tak nol diberlakukan untuk menghindari kasus yang tidak penting seperti $A0=0$, yang mana berlaku stiap A an 0

- $Ax$: vektor x setelahtransformasi linier
- $\lambda$: perkalian bilangan real dari X,i.e peregangan vektor sebanyak nilai eigen kali.
  
misalkan kita memiliki matriks $A$ sebagai berikut:  

$$
	A= \begin{bmatrix} 
	2 & 1 \\
	1 & 2 \\
	\end{bmatrix}
$$

**mencari eigen value**  
untuk mencari nilai eigen , kita harus menyelesaikan determinan dari $A-\lambda I$ dimana $\lambda$ adalah nilai eigen dan I adalah matriks identitas  

$A-\lambda = |2-\lambda\\1 2-\lambda|$ = $(2-\lambda) -(1)(1) = (2-\lambda)^2$

maka, persamaan karakteristiknya adalah:
$$ (2-\lambda)^2 -1 = 0\\
(2-\lambda)^2 = 1\\
2-\lambda = +- 1\\
\lambda 1 = 1, \lambda 2 = 3
$$

jadi, nilai eigennya adalah $\lambda$ = 1 dan $\lambda2 = 3$

**mencari eigen vector**  
untuk $\lambda1=1$

untuk $\lambda 1=1$ kita subtitusikan $\lambda 1=1$ ke dalam persamaan $A-\lambda I$ :
$$
A-1 I=\left[\begin{array}{cc}
2-1 & 1 \\
1 & 2-1
\end{array}\right]=\left[\begin{array}{ll}
1 & 1 \\
1 & 1
\end{array}\right]
$$
kita mencari vector $\mathrm{v}=\left[\begin{array}{l}x \\ y\end{array}\right]$ sehingga: $\left[\begin{array}{ll}1 & 1 \\ 1 & 1\end{array}\right]\left[\begin{array}{l}x \\ y\end{array}\right]=\left[\begin{array}{l}
0 \\
0
\end{array}\right]$

jadi,vector eigennya untuk \lambda $1=1$ adalah v1 $=\left[\begin{array}{c}1 \\ -1\end{array}\right]$.
untuk $\lambda 2=3$ :
kita subtitusikan $\lambda 2=3$ ke dalam persamaan A - $\lambda I$ :

$$
=\left[\begin{array}{cc}
-1 & 1 \\
1 & -1
\end{array}\right] \quad A-3 I=\left[\begin{array}{cc}
2-3 & 1 \\
1 & 2-3
\end{array}\right]
$$
kita mencari vektor $\left[\begin{array}{l}x \\ y\end{array}\right]$ sehingga:
$$
\left[\begin{array}{cc}
-1 & 1 \\
1 & -1
\end{array}\right]\left[\begin{array}{l}
x \\
y
\end{array}\right]=\left[\begin{array}{l}
0 \\
0
\end{array}\right]
$$
jadi, vector eigennya untuk $\lambda 2=3$ adalahv $2= \left[\begin{array}{l}
1 \\
1
\end{array}\right]$




```python
# Mengimpor library yang diperlukan
import numpy as np

# Definisikan matriks A
A = np.array([[2, 1],
              [1, 2]])

# Menghitung nilai eigen dan vektor eigen
eigenvalues, eigenvectors = np.linalg.eig(A)

# Menampilkan hasil
print("Matriks A:")
print(A)

print("\nNilai Eigen:")
print(eigenvalues)

print("\nVektor Eigen:")
print(eigenvectors)
```
