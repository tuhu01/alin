# Matriks Refleksi
Matriks refleksi adalah matriks yang digunakan dalam matematika, khususnya dalam geometri dan aljabar linear, untuk merepresentasikan transformasi refleksi suatu vektor terhadap suatu garis atau bidang. Refleksi ini berarti bahwa setiap titik dari suatu objek dipetakan ke titik lain di sisi yang berlawanan dari garis atau bidang tersebut, seolah-olah objek tersebut dipantulkan oleh sebuah cermin.

Berikut adalah beberapa contoh matriks refleksi dalam dua dimensi (2D) dan tiga dimensi (3D):

# Refleksi dalam 2D
1. **Refleksi terhadap sumbu-x**: 
   \[
   \begin{bmatrix}
   1 & 0 \\
   0 & -1
   \end{bmatrix}
   \]
   Matriks ini merefleksikan titik (x, y) menjadi (x, -y).

2. **Refleksi terhadap sumbu-y**: 
   \[
   \begin{bmatrix}
   -1 & 0 \\
   0 & 1
   \end{bmatrix}
   \]
   Matriks ini merefleksikan titik (x, y) menjadi (-x, y).

3. **Refleksi terhadap garis y = x**:
   \[
   \begin{bmatrix}
   0 & 1 \\
   1 & 0
   \end{bmatrix}
   \]
   Matriks ini merefleksikan titik (x, y) menjadi (y, x).

### Refleksi dalam 3D
1. **Refleksi terhadap bidang xy**:
   \[
   \begin{bmatrix}
   1 & 0 & 0 \\
   0 & 1 & 0 \\
   0 & 0 & -1
   \end{bmatrix}
   \]
   Matriks ini merefleksikan titik (x, y, z) menjadi (x, y, -z).

2. **Refleksi terhadap bidang yz**:
   \[
   \begin{bmatrix}
   -1 & 0 & 0 \\
   0 & 1 & 0 \\
   0 & 0 & 1
   \end{bmatrix}
   \]
   Matriks ini merefleksikan titik (x, y, z) menjadi (-x, y, z).

3. **Refleksi terhadap bidang zx**:
   \[
   \begin{bmatrix}
   1 & 0 & 0 \\
   0 & -1 & 0 \\
   0 & 0 & 1
   \end{bmatrix}
   \]
   Matriks ini merefleksikan titik (x, y, z) menjadi (x, -y, z).

Untuk refleksi terhadap garis atau bidang yang tidak sejajar dengan sumbu koordinat, matriks refleksi dapat disusun menggunakan teknik transformasi yang melibatkan rotasi dan skala. Matematika di balik penyusunan matriks ini lebih kompleks dan memerlukan pemahaman yang lebih mendalam tentang transformasi linear dan koordinat homogen.

Secara umum, matriks refleksi adalah alat yang sangat penting dalam berbagai bidang seperti grafika komputer, fisika, dan robotika, di mana transformasi geometris sering digunakan untuk memanipulasi dan memahami objek dalam ruang.
